```python
# Checking for Missing Values 

attrition_data.isnull().sum()

# There are no missing values in the dataset. Hence data can be prepared by converting categorical to dummy variables to include them in 

#the model.
```

```python
# Transforming categorical variables to dummy variables 
#Business Travel

bus_travel = pd.get_dummies(attrition_data['BusinessTravel'],drop_first=True)

bus_travel
```

```python
#Department

dept= pd.get_dummies(attrition_data['Department'],drop_first=True)

dept
```

```python
#Education Field

edu_field= pd.get_dummies(attrition_data['EducationField'],drop_first=True)
edu_field
```

```python
# Gender 

Gender= pd.get_dummies(attrition_data['Gender'],drop_first=True)
Gender
```

```python
# Job Role

jobrole= pd.get_dummies(attrition_data['JobRole'],drop_first=True)
jobrole
```

```python
mar_status= pd.get_dummies(attrition_data['MaritalStatus'],drop_first=True)
mar_status
```

```python
#Over18
over18= pd.get_dummies(attrition_data['Over18'])
over18
```

```python
over_time= pd.get_dummies(attrition_data['OverTime'],drop_first=True)
over_time
```

```python
#Concatenating all dummies 

attrition_data = pd.concat([attrition_data,bus_travel,dept,edu_field,Gender,jobrole,mar_status,over18,over_time],axis=1)
```

```python
attrition_data.head(2)
```

```python
attrition_data.columns
```

```python
attrition_data.drop(['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime'],axis=1,inplace=True)
```

```python
attrition_data.head(2)
```

# Preparing Training and Testing Data

```python
#Assigning X and Y for Train Data 

X= attrition_data.drop("Attrition",axis=1)
y = attrition_data["Attrition"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
# Training the logistic Regression Model

lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```
