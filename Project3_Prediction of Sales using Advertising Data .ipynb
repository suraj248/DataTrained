# Prediction of Sales Using Linear Regression

```python
#Importing the packages pandas,Numpy,SeaBorn,matplotlib,math and Numpy 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
%matplotlib inline

#Importing the Advertsing_Dataset

Adv_data = pd.read_csv('C:/Users/Suraj/Jupyter Projects/Project 3- Advertising Project/Advertising.csv')
Adv_data.head()

#We can see that dataset contains 4 columns and 200 rows
```

```python
#Using seabornplot to analyse the scatterplot of the 3 columns with the Target Variable Sales.

sns.pairplot(Adv_data, x_vars=['TV', 'radio', 'newspaper'], y_vars='sales',height=7, aspect=0.7, kind='reg')
```

```python
#There seems to be a strong association between TV v/s Sales in comparison with Radio v/s Sales.

# Association seems to be way too weak with Newspaper v/s Sales.

#There are only numeric variables and can see a linear trend with the variables .
# Hence algorithm to be used would be linear regression.
```

```python
# Preparing Data for Modeling 

#We are choosing the 3 variables as independent and dropping the Sales Variable since it will be our Target variable. 

X = Adv_data[['TV','radio','newspaper']]

y = Adv_data["sales"]
```

```python
#Splitting Data into Train and Test 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
#Importing Logistic Regression Model

from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
#Training the Data

lr.fit(X_train,y_train)

lr.score(X_train,y_train)

#The trainingdata R square value was found to be 92.1%
```

```python
pred=lr.predict(X_test)
```

```python
# Interpreting Model Coefficients

print(lr.intercept_)
print(lr.coef_)

#The intercept is 3.4106415886118224 and the coefficients are  0.04303172 ,0.19352212 and -0.00386729
```

```python
lr.score(X_test,y_test)

#R-square value of Test data was found to be 86%.
```

```python
#Plot for Residual Error

#Setting ploting style

plt.style.use('fivethirtyeight')

#Plotting Residual errors in training Data

plt.scatter(lr.predict(X_train), lr.predict(X_train) - y_train, 
            color = "green", s = 15, label = 'Train data')
```

```python
# Plotting Residual errors in Test Data 

plt.scatter(lr.predict(X_test), lr.predict(X_test) - y_test, 
            color = "blue", s = 15, label = 'Test data') 
```

```python
#Summary of R-Squares 

# R-square value with Train Data - 92.10% 

#R-square with Test Data -  86 %
```
