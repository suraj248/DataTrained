```python
churn_data=pd.concat([churn_data,gen,partner,dep,p_serv,mul_lines,i_serv,o_sec,o_backup,d_protect,T_sup,stv,smovies,contract,p_bill,p_method],axis=1)

churn_data.head(2)
```

```python
# Dropping original variables

churn_data.drop(['gender','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod'],axis=1, inplace=True)
```

```python
churn_data.head(2)
```

```python
# Assigning X and Y 

X= churn_data.drop("Churn",axis=1)
y = churn_data["Churn"]

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45)
```

```python
#Importing Logistic Regression from Scikit Learn

from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
```

```python
# Training the model

lr.fit(X_train,y_train)
```

```python
# Predicting with test dataset 

pred = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,pred)

# Precision is 65%

# Recall is 58%

# Accuracy is 74%
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,pred)
```

```python
from sklearn.metrics import accuracy_score

from sklearn import metrics
```

```python
accuracy_score(y_test,pred)

# Accuracy of the model is 81%
```

```python
print("Accuracy", metrics.accuracy_score(y_test, pred))
```

```python
# AUC Curve 

y_pred_proba = lr.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
